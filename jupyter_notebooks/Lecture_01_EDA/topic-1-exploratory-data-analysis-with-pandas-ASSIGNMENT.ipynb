```python
import numpy as np
import pandas as pd
pd.set_option('display.max.columns', 100)
pd.set_option('display.max_rows', None)  
# to draw pictures in jupyter notebook
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns
# we don't like warnings
# you can comment the following 2 lines if you'd like to
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../../data/adult.data.csv')
df.head()
```

**1. How many men and women (*sex* feature) are represented in this dataset?** 

```python
pd.DataFrame(df['sex'].value_counts())
```

**2. What is the average age (*age* feature) of women?**

```python
df.loc[df['sex'] == 'Female',['age']].mean()
```

**3. What is the percentage of German citizens (*native-country* feature)?**

```python
print("The percentage of 'German' citizens in the dataset is:\n {0}%"
      .format(round(100*len(df.loc[df['native-country'] == 'Germany']) 
                    / len(df), 2)))
```

**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year? **

```python
df.groupby(['salary'])['age'].describe()[['mean','std']]
```

**6. Is it true that people who earn more than 50K have at least high school education? (*education â€“ Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* or *Doctorate* feature)**

```python
# NO
df.loc[df['salary'] == '>50K']['education'].unique()
```

**7. Display age statistics for each race (*race* feature) and each gender (*sex* feature). Use *groupby()* and *describe()*. Find the maximum age of men of *Amer-Indian-Eskimo* race.**

```python
# Age statistics for each race
df.groupby(['race','sex'])[['age']].describe()
```

```python
# Maximum age of men in Amer-Indian-Eskimo race
df.loc[(df['race'] == 'Amer-Indian-Eskimo') & (df['sex'] == 'Male')].groupby(['race','sex'])[['age']].max()
```

**8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
df['marital-status'].value_counts()
```

```python
# using the capitalized 'Married' as a search term so we can distinguish between married and singles

print("Proportion of married and earning >50K: \n",
      len(df[df['marital-status'].str.contains('Married')].loc[df['salary'] == '>50K'][df['sex'] == 'Male']))

print("Proportion of single and earning >50K: \n",
      len(df[~df['marital-status'].str.contains('Married')].loc[df['salary'] == '>50K'][df['sex'] == 'Male']))
```

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
print("The maximum number of hours a person work per week is:\n" ,df['hours-per-week'].max())

print("The number of people who work maximum number of hours:\n" ,len(df[df['hours-per-week'] == df['hours-per-week'].max()]))

print("The percentage of people who work the maximum amount of hours per week and earn >50K is: \n {0}%"
      .format(round(100*len(df.loc[(df['salary'] == '>50K') 
           & (df['hours-per-week'] == df['hours-per-week'].max())]) / 
                    len(df[df['hours-per-week'] == df['hours-per-week'].max()]), 2)))
# print(len(df.loc[(df['salary'] == '>50K') & (df["hours-per-week"] == 99)]) / len(df.loc[df['hours-per-week'] == 99]))
```

**10. Count the average time of work (*hours-per-week*) for those who earn a little and a lot (*salary*) for each country (*native-country*). What will these be for Japan?**

```python
# all countires
df.groupby(['native-country', 'salary'])[['hours-per-week']].mean()
```

```python
# Japan only
df[df['native-country'] == 'Japan'][['salary','native-country','hours-per-week']].groupby(['native-country','salary']).mean()
```
